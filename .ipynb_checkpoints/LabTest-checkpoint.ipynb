```python
a = int(input("Enter a number: "))

for i in range(1, a+1):
    prime = 0
    for j in range(1, i+1):
        if(i % j == 0):
            prime += 1;

    if(prime == 2):
        print(i)
```

```python
a = int(input("Enter a number: "))
b = int(input("Enter a number: "))


gcd=1
for i in range(1, min(a, b)+1):
    if(a%i == 0 and b%i == 0):
        gcd = i

print(gcd)
```
