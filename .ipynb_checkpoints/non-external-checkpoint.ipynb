### Stack 

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.stack.len == 0:
            print("Stack Empty")
            return None
        else:
            return self.stack.pop()

    def top(self):
        if self.stack.len == 0:
            print("Stack Empty")
            return None
        else:
            return self.stack[-1]

    def display(self):
        for i in self.stack:
            print(i)

stack = Stack()

stack.push(3)
stack.push(3)
stack.push(3)
stack.push(3)
stack.push(3)
stack.push(3)
```

## Queue

```python
class Queue:
    def __init__(self, max_size):
        self.queue = []
        self.max = max_size

    def enqueue(self, add_item):
        if len(self.queue) >= self.max:
            print("Queue is Full")
        else:
            self.queue.append(add_item)

    def dequeue(self):
        if not self.queue:
            print("Queue Underflow")
            return None
        else:
            pop_item = self.queue.pop(0)
            return pop_item

    def display(self):
        if not self.queue:
            print("Queue is empty")
        else:
            for i in self.queue:
                print(i, end= " ")
            print()

queue = Queue(max_size=5)
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.enqueue(4)
queue.enqueue(5)
queue.enqueue(6)

queue.display()

queue.dequeue()
queue.dequeue()


queue.display()

```

## Graph

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_nodes(self, node):
        if node not in self.graph:
            self.graph[node] = []

    def add_edge(self, edge1, edge2):
        self.add_nodes(edge1)
        self.add_nodes(edge2)
        self.graph[edge1].append(edge2)
        self.graph[edge2].append(edge1)

    def get_neighbours(self, node):
        return self.graph.get(node, [])

    def display(self):
        for nodes, neighbours in self.graph.items():
            print(f"{nodes} -> {neighbours} \n")

graph = Graph()
graph.add_edge('A', 'B')
graph.add_edge('A', 'C')
graph.add_edge('B', 'D')
graph.add_edge('C', 'D')
graph.add_nodes('E')
graph.display()

graph.get_neighbours('B')
```
